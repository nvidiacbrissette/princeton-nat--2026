# Getting Started with NeMo Agent Toolkit CLI

## What is the NVIDIA NeMo Agent Toolkit?

The NVIDIA NeMo Agent Toolkit is a open-source, lightweight library developed by NVIDIA for composing and configuring agents of any framework. It provides MLOps administrators, developers, data solutions engineers, and data scientists with the tools to create workflows that are repeatable and flexible.

Agentic system developers can interact with the NeMo Agent Toolkit using a workflow instance that consists of three primary interfaces, which are configurable using a YAML file:

- **Workflow**: The main entry point of a NeMo Agent Toolkit application that orchestrates the agentic system, composed of an arbitrary number of agents and tools across a heterogeneous set of implementations.

- **Tool**: The NeMo Agent Toolkit primitive tools provided to the workflows that agents use or access. They are arbitrarily complex coroutines containing skill implementations that get wrapped in specified LLM frameworks at runtime.

- **LLM**: The core reasoning engines for NeMo Agent Toolkit workflows and can be used to route requests to the appropriate tool. Each LLM comes from a provider (NIM, OpenAI, etc.) and gets wrapped as an LLM framework client at runtime.

The NeMo Agent Toolkit can run agent workflows on the command line, serve them as an extensible API, evaluate agents for testing, run deep profiling, and provide observability into the entire agentic pipeline.
