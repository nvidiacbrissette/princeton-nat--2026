```python
%%writefile ../.env
NVIDIA_API_KEY=paste-your-key-here
TAVILY_API_KEY=paste-your-key-here
```

Run the cell below to verify your API keys are working correctly.

```python
import os
import requests
from dotenv import find_dotenv, load_dotenv

load_dotenv(find_dotenv())

# Test NVIDIA API Key
nvidia_key = os.environ.get("NVIDIA_API_KEY", "")
assert nvidia_key and nvidia_key != "paste-your-key-here", (
    "NVIDIA_API_KEY is not set. Edit the %%writefile cell above with your key and rerun it."
)
response = requests.post(
    "https://integrate.api.nvidia.com/v1/chat/completions",
    headers={"Authorization": f"Bearer {nvidia_key}"},
    json={
        "model": "meta/llama-3.1-70b-instruct",
        "messages": [{"role": "user", "content": "Say 'hello' and nothing else."}],
        "max_tokens": 5,
    },
)
assert response.status_code == 200, (
    f"NVIDIA API request failed: {response.status_code} {response.text}"
)
reply = response.json()["choices"][0]["message"]["content"].strip()
print(f"NVIDIA_API_KEY: OK  (response: {reply})")

# Test Tavily API Key
tavily_key = os.environ.get("TAVILY_API_KEY", "")
assert tavily_key and tavily_key != "paste-your-key-here", (
    "TAVILY_API_KEY is not set. Edit the %%writefile cell above with your key and rerun it."
)
response = requests.post(
    "https://api.tavily.com/search",
    json={"query": "NVIDIA", "max_results": 1, "api_key": tavily_key},
)
assert response.status_code == 200, (
    f"Tavily API request failed: {response.status_code} {response.text}"
)
print(f"TAVILY_API_KEY: OK  (got {len(response.json().get('results', []))} result(s))")

print("\nAll API keys validated successfully!")
```
